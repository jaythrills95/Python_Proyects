```python
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.sort_values('job_posted_date', inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
date_counts = df.job_posted_month.value_counts()
date_counts = date_counts.sort_index()
date_counts
```

```python
plt.plot(date_counts.index, date_counts.values)
```

```python
job_counts = df.job_title_short.value_counts()
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts.values)
plt.title('Posting by Job Title')
plt.ylabel('Count of job Postings')
plt.xticks(rotation=45)
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.xticks(rotation=45, ha='right')
plt.title('Posting by Job Title')
plt.show()
```

```python
job_counts.plot(kind='bar', width=0.8)
colors = ['blue' if i < len(job_counts) - 3 else 'red' for i in range(len(job_counts))]
plt.bar(job_counts.index, job_counts.values, color=colors)
plt.bar(job_counts.index, job_counts.values, color=colors)
plt.xticks(rotation=45, ha='right')
plt.title('Posting by Job Title')
plt.show()
```

```python
mean_salaries = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
colors = ['blue' if i < len(mean_salaries) - 3 else 'red' for i in range(len(mean_salaries))]
mean_salaries.plot(kind='barh', width=0.8, color=colors)
plt.xlabel('Mean Salary')
plt.ylabel('Job Title')
plt.title('Median Salary by Job Title')
plt.show()
```
